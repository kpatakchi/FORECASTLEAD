```python
import matplotlib.pyplot as plt

# Step frequency for 12 UTC
forecast_steps = list(range(0, 91)) + list(range(93, 145, 3)) + list(range(150, 241, 6))

# Time intervals for each step (assuming each step corresponds to 1 hour)
time_intervals = [i for i in range(len(forecast_steps))]

# Plotting the forecast steps against time intervals
plt.figure(figsize=(8, 6), dpi=200)
plt.plot(time_intervals, forecast_steps,'o', color='blue', ms=2)
plt.title('HRES Forecast Steps - 12 UTC')
plt.xlabel('Time Intervals (hours)')
plt.ylim(0, 241)
plt.ylabel('Forecast Step')
plt.grid(True)
plt.show()
```
